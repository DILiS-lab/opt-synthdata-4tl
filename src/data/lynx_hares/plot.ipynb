```python
import os

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.font_manager as fm
from pathlib import Path
import matplotlib as mpl


# Define your font
fpath = Path(mpl.get_data_path(), "fonts/ttf/cmr10.ttf")
fontprop = fm.FontProperties(fname=fpath, size=14)
fontprop_small = fm.FontProperties(fname=fpath, size=14)
COLORS=["#888888", "#B1053A", "#6699CC", "#44AA99", "#CC6677"]
```

```python
url = 'http://people.whitman.edu/~hundledr/courses/M250F03/LynxHare.txt'
lynx_hares = pd.read_csv(url, delim_whitespace=True, header=None, index_col=0)
lynx_hares.index.name = 'year'
lynx_hares.columns = ['hare', 'lynx']
lynx_hares
```

```python
TEST_SPLIT = 0.80
train_test_split_year_0_2 = round(min(lynx_hares.index) + (1-TEST_SPLIT) * (max(lynx_hares.index) - min(lynx_hares.index)))
train_test_split_year_0_2

# Set figsize
sns.set_style("whitegrid")
plt.figure(figsize=(10, 4))

# Create a scatter plot with connected lines using Seaborn
sns.lineplot(x='year', y='lynx', data=lynx_hares, marker='^', label='Lynx', color=COLORS[1], linewidth=2, markersize=6)
sns.lineplot(x='year', y='hare', data=lynx_hares, marker='o', label='Hare', color=COLORS[0], linewidth=2, markersize=6)

plt.axvline(x = round(train_test_split_year_0_2), color = 'black', linestyle = '-', linewidth = '1.0')
plt.text(round(train_test_split_year_0_2)-2.5, 100, 'train cutoff', rotation=90, color = 'black', fontsize=14, font=fpath)

# Set plot title and labels
plt.legend(prop=fontprop)
plt.xticks(fontproperties=fontprop)
plt.yticks(fontproperties=fontprop)
plt.xlabel('Years', fontproperties=fontprop)
plt.ylabel('Value', fontproperties=fontprop)
```
